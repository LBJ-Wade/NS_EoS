```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc
```

## Plotting adjustments

```python
%matplotlib inline

# Set `usetex=False' if you do not have LaTeX installed.
rc('text', usetex=True)
rc('font', family='serif')
mpl.rcParams['text.latex.preamble'] = [r"\usepackage{amsmath}"]
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
width = 3.4 * 2
golden = 1.61803398875
height = width / golden
```

```python
colours = ["#d32f2f", "#7b1fa2", "#303f9f", "#0097a7", "#689f38", "#ffa000", "#B8B8B8", "#484848"]
lines = [[], [1, 2], [2, 1], [4, 1], [8, 1], [4, 1, 2, 1]]
```

## Equilibrium composition

```python
data_LNS = pd.read_csv("../examples/data/parameters_LNS.txt")
data_NRAPR = pd.read_csv("../examples/data/parameters_NRAPR.txt")
data_Skchi450 = pd.read_csv("../examples/data/parameters_Skchi450.txt")
data_SLy4 = pd.read_csv("../examples/data/parameters_SLy4.txt")
data_SQMC700 = pd.read_csv("../examples/data/parameters_SQMC700.txt")
data_Ska35s20 = pd.read_csv("../examples/data/parameters_Ska35s20.txt")
```

```python
data_LNS = data_LNS[1:]
data_NRAPR = data_NRAPR[1:]
data_Skchi450 = data_Skchi450[1:]
data_SLy4 = data_SLy4[1:]
data_SQMC700 = data_SQMC700[1:]
data_Ska35s20 = data_Ska35s20[1:]
```

```python
data_NRAPR.head()
```

```python
n_b = pd.to_numeric(data_LNS["n_b"]).values
rho_b_log = np.log10(pd.to_numeric(data_LNS["rho_b"]).values)
rho_b = pd.to_numeric(data_LNS["rho_b"]).values
```

```python
x_p_LNS = pd.to_numeric(data_LNS["n_p"]).values / n_b
x_p_NRAPR = pd.to_numeric(data_NRAPR["n_p"]).values / n_b
x_p_Skchi450 = pd.to_numeric(data_Skchi450["n_p"]).values / n_b
x_p_SLy4 = pd.to_numeric(data_SLy4["n_p"]).values / n_b
x_p_SQMC700 = pd.to_numeric(data_SQMC700["n_p"]).values / n_b
x_p_Ska35s20 = pd.to_numeric(data_Ska35s20["n_p"]).values / n_b
```

```python
x_n_LNS = pd.to_numeric(data_LNS["n_n"]).values / n_b
x_n_NRAPR = pd.to_numeric(data_NRAPR["n_n"]).values / n_b
x_n_Skchi450 = pd.to_numeric(data_Skchi450["n_n"]).values / n_b
x_n_SLy4 = pd.to_numeric(data_SLy4["n_n"]).values / n_b
x_n_SQMC700 = pd.to_numeric(data_SQMC700["n_n"]).values / n_b
x_n_Ska35s20 = pd.to_numeric(data_Ska35s20["n_n"]).values / n_b
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(rho_b * 1e-14, x_p_LNS, '-', color=colours[0], linewidth=3, label=r"LNS", dashes=lines[0])
ax.plot(rho_b * 1e-14, x_p_NRAPR, '-', color=colours[1], linewidth=3, label=r"NRAPR", dashes=lines[1])
ax.plot(rho_b * 1e-14, x_p_Skchi450, '-', color=colours[2], linewidth=3, label=r"Sk$\chi$450", dashes=lines[2])
ax.plot(rho_b * 1e-14, x_p_SLy4, '-', color=colours[3], linewidth=3, label=r"SLy4", dashes=lines[3])
ax.plot(rho_b * 1e-14, x_p_SQMC700, '-', color=colours[4], linewidth=3, label=r"SQMC700", dashes=lines[4])
ax.plot(rho_b * 1e-14, x_p_Ska35s20, '-', color=colours[5], linewidth=3, label=r"Ska35s20", dashes=lines[5])

ax.set_xlim(1., 10)
ax.set_ylim(0.015, 0.11)
ax.set_xlabel(r"$\rho$ $\left[ 10^{14} \, {\rm g} / {\rm cm}^{3} \right]$")
ax.set_ylabel(r"$x_{\rm p}$")

plt.legend(loc=4, handlelength=3.2)

plt.savefig('../examples/plots/proton_fractions.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(rho_b * 1e-14, x_n_LNS, '-', color=colours[0], linewidth=3, label=r"LNS", dashes=lines[0])
ax.plot(rho_b * 1e-14, x_n_NRAPR, '-', color=colours[1], linewidth=3, label=r"NRAPR", dashes=lines[1])
ax.plot(rho_b * 1e-14, x_n_Skchi450, '-', color=colours[2], linewidth=3, label=r"Sk$\chi$450", dashes=lines[2])
ax.plot(rho_b * 1e-14, x_n_SLy4, '-', color=colours[3], linewidth=3, label=r"SLy4", dashes=lines[3])
ax.plot(rho_b * 1e-14, x_n_SQMC700, '-', color=colours[4], linewidth=3, label=r"SQMC700", dashes=lines[4])
ax.plot(rho_b * 1e-14, x_n_Ska35s20, '-', color=colours[5], linewidth=3, label=r"Ska35s20", dashes=lines[5])

ax.set_xlim(1., 10)
ax.set_ylim(0.89, 0.985)
ax.set_xlabel(r"$\rho$ $\left[ 10^{14} \, {\rm g} / {\rm cm}^{3} \right]$")
ax.set_ylabel(r"$x_{\rm n}$")

plt.legend(loc=1, handlelength=3.2)

plt.savefig('../examples/plots/neutron_fractions.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

## Superfluid parameters

```python
data_gaps = pd.read_csv("../examples/data/energy_gaps.txt")
data_gaps = data_gaps[1:]
```

```python
data_gaps.head()
```

### Gaps as function of Fermi number

```python
k_F = pd.to_numeric(data_gaps["k_F"]).values
Delta_p = pd.to_numeric(data_gaps["Delta_p"]).values
Delta_n = pd.to_numeric(data_gaps["Delta_n"]).values
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(k_F, Delta_p, '-', color=colours[6], linewidth=3, label=r"$\Delta_{\rm p}$ (CCDK)", dashes=lines[0])
ax.plot(k_F, Delta_n, '-', color=colours[7], linewidth=3, label=r"$\Delta_{\rm n}$ (TToa)", dashes=lines[1])

ax.set_xlim(0., 3.2)
ax.set_ylim(0.0, 1.2)
ax.set_xlabel(r"$k_{\rm F p,n}$ [${\rm 1/fm}$]")
ax.set_ylabel(r"$\Delta$ [${\rm MeV}$]")

plt.legend(loc=1, handlelength=3.2)

plt.savefig('../examples/plots/energy_gaps.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

### Gaps as function of density

```python
Delta_p_NRAPR = pd.to_numeric(data_NRAPR["Delta_p"]).values
Delta_n_NRAPR = pd.to_numeric(data_NRAPR["Delta_n"]).values
Delta_p_Skchi450 = pd.to_numeric(data_Skchi450["Delta_p"]).values
Delta_n_Skchi450 = pd.to_numeric(data_Skchi450["Delta_n"]).values
Delta_p_SLy4 = pd.to_numeric(data_SLy4["Delta_p"]).values
Delta_n_SLy4 = pd.to_numeric(data_SLy4["Delta_n"]).values
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(rho_b * 1e-14, Delta_p_NRAPR, '-', color=colours[1], linewidth=3, label=r"$\Delta_{\rm p}$ NRAPR", dashes=lines[0])
ax.plot(rho_b * 1e-14, Delta_n_NRAPR, '-', color=colours[1], linewidth=3, label=r"$\Delta_{\rm n}$ NRAPR", dashes=lines[1])
ax.plot(rho_b * 1e-14, Delta_p_Skchi450, '-', color=colours[2], linewidth=3, label=r"$\Delta_{\rm p}$ Sk$\chi450$", dashes=lines[0])
ax.plot(rho_b * 1e-14, Delta_n_Skchi450, '-', color=colours[2], linewidth=3, label=r"$\Delta_{\rm n}$ Sk$\chi450$", dashes=lines[1])
ax.plot(rho_b * 1e-14, Delta_p_SLy4, '-', color=colours[3], linewidth=3, label=r"$\Delta_{\rm p}$ SLy4", dashes=lines[0])
ax.plot(rho_b * 1e-14, Delta_n_SLy4, '-', color=colours[3], linewidth=3, label=r"$\Delta_{\rm n}$ SLy4", dashes=lines[1])

ax.set_xlim(1., 10)
ax.set_ylim(0.0, 1.2)
ax.set_xlabel(r"$\rho$ $\left[ 10^{14} \, {\rm g} / {\rm cm}^{3} \right]$")
ax.set_ylabel(r"$\Delta$ [${\rm MeV}$]")

plt.legend(loc=3, handlelength=3.2)

plt.savefig('../examples/plots/energy_gaps_density.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

### Temperature as function of density

```python
k_B = 8.617330 * 1e-11 # [MeV/K] 
c_singlet = 0.567
c_triplet = 0.118

Tc_p_NRAPR = c_singlet * Delta_p_NRAPR / k_B
Tc_n_NRAPR = c_triplet * Delta_n_NRAPR / k_B
Tc_p_Skchi450 = c_singlet * Delta_p_Skchi450/ k_B
Tc_n_Skchi450 = c_triplet * Delta_n_Skchi450 / k_B
Tc_p_SLy4 = c_singlet * Delta_p_SLy4 / k_B
Tc_n_SLy4 = c_triplet * Delta_n_SLy4 / k_B
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(rho_b * 1e-14, Tc_p_NRAPR * 1e-9, '-', color=colours[1], linewidth=3, label=r"$T_{\rm cp}$ NRAPR", dashes=lines[0])
ax.plot(rho_b * 1e-14, Tc_n_NRAPR * 1e-9, '-', color=colours[1], linewidth=3, label=r"$T_{\rm cn}$ NRAPR", dashes=lines[1])
ax.plot(rho_b * 1e-14, Tc_p_Skchi450 * 1e-9, '-', color=colours[2], linewidth=3, label=r"$T_{\rm cp}$ Sk$\chi450$", dashes=lines[0])
ax.plot(rho_b * 1e-14, Tc_n_Skchi450 * 1e-9, '-', color=colours[2], linewidth=3, label=r"$T_{\rm cn}$ Sk$\chi450$", dashes=lines[1])
ax.plot(rho_b * 1e-14, Tc_p_SLy4 * 1e-9, '-', color=colours[3], linewidth=3, label=r"$T_{\rm cp}$ SLy4", dashes=lines[0])
ax.plot(rho_b * 1e-14, Tc_n_SLy4 * 1e-9, '-', color=colours[3], linewidth=3, label=r"$T_{\rm cn}$ SLy4", dashes=lines[1])

ax.set_xlim(1., 10.)
ax.set_ylim(0.0, 7)
ax.set_xlabel(r"$\rho$ $\left[ 10^{14} \, {\rm g} / {\rm cm}^{3} \right]$")
ax.set_ylabel(r"$T_{\rm c}$ [$10^9 \, {\rm K}$]")

plt.legend(loc=3, handlelength=3.2)

plt.savefig('../examples/plots/Tc_density.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

### Coherence lengths

```python
xi_p_NRAPR = pd.to_numeric(data_NRAPR["xi_p"]).values
xi_n_NRAPR = pd.to_numeric(data_NRAPR["xi_n"]).values
xi_p_Skchi450 = pd.to_numeric(data_Skchi450["xi_p"]).values
xi_n_Skchi450 = pd.to_numeric(data_Skchi450["xi_n"]).values
xi_p_SLy4 = pd.to_numeric(data_SLy4["xi_p"]).values
xi_n_SLy4 = pd.to_numeric(data_SLy4["xi_n"]).values
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(width, height)

ax.plot(rho_b * 1e-14, xi_p_NRAPR * 1e11, '-', color=colours[1], linewidth=3, label=r"$\xi_{\rm p}$ NRAPR", dashes=lines[0])
ax.plot(rho_b * 1e-14, xi_n_NRAPR * 1e11, '-', color=colours[1], linewidth=3, label=r"$\xi_{\rm n}$ NRAPR", dashes=lines[1])
ax.plot(rho_b * 1e-14, xi_p_Skchi450 * 1e11, '-', color=colours[2], linewidth=3, label=r"$\xi_{\rm p}$ Sk$\chi450$", dashes=lines[0])
ax.plot(rho_b * 1e-14, xi_n_Skchi450 * 1e11, '-', color=colours[2], linewidth=3, label=r"$\xi_{\rm n}$ Sk$\chi450$", dashes=lines[1])
ax.plot(rho_b * 1e-14, xi_p_SLy4 * 1e11, '-', color=colours[3], linewidth=3, label=r"$\xi_{\rm p}$ SLy4", dashes=lines[0])
ax.plot(rho_b * 1e-14, xi_n_SLy4 * 1e11, '-', color=colours[3], linewidth=3, label=r"$\xi_{\rm n}$ SLy4", dashes=lines[1])

ax.set_xlim(1., 10.)
ax.set_ylim(0.0, 10)
ax.set_xlabel(r"$\rho$ $\left[ 10^{14} \, {\rm g} / {\rm cm}^{3} \right]$")
ax.set_ylabel(r"$\xi$ [$10^{-11} \, {\rm cm}$]")

plt.legend(loc=2, handlelength=3.2)

plt.savefig('../examples/plots/coherence_lengths_density.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```
